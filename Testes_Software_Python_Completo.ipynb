### Conceitos Teóricos
**Teste de Software** é o processo de verificar se um sistema ou componente funciona corretamente. Os principais tipos são:

| Tipo de Teste | O que testa? | Quando é feito? |
|---------------|--------------|-----------------|
| **Unitário** | Uma função isolada | Durante o desenvolvimento |
| **Integração** | A interação entre funções | Após os testes unitários |
| **Sistema** | O sistema como um todo | Antes da entrega ao usuário |

````python
# 📦 Módulo conta.py
# Classe que representa uma conta bancária
class Conta:
    def __init__(self, titular, saldo=0):
        self.titular = titular  # Nome do titular da conta
        self.saldo = saldo      # Saldo inicial da conta

    def consultar_saldo(self):
        # Retorna o saldo atual da conta
        return self.saldo
````

````python
# 📦 Módulo transacoes.py
# Função para depositar um valor na conta
def depositar(conta, valor):
    if valor <= 0:
        raise ValueError("Valor de depósito deve ser positivo")
    conta.saldo += valor  # Atualiza o saldo
    return conta.saldo

# Função para sacar um valor da conta
def sacar(conta, valor):
    if valor <= 0:
      raise ValueError("O Valor do saldo deve ser maior que zero ")
    if valor > conta.saldo:
        raise ValueError("Saldo insuficiente")
    conta.saldo -= valor  # Atualiza o saldo
    return conta.saldo
````

````python
# ✅ Testes Unitários com unittest
import unittest

# Classe de testes para as funções de transação
class TestTransacoes(unittest.TestCase):
    def setUp(self):
        self.conta = Conta("João", 100)

    def test_deposito(self):
        novo_saldo = depositar(self.conta, 50)
        self.assertEqual(novo_saldo, 150)

    def test_saque(self):
        novo_saldo = sacar(self.conta, 40)
        self.assertEqual(novo_saldo, 60)

    def test_saque_invalido(self):
        with self.assertRaises(ValueError):
            sacar(self.conta, 200)

    def test_deposito_invalido(self):
        with self.assertRaises(ValueError):
            depositar(self.conta, -10)

    def test_saque_valor_negativo(self):
        with self.assertRaises(ValueError):
            sacar(self.conta, -50)

    def test_saque_valor_zero (self):
        with self.assertRaises(ValueError):
            sacar(self.conta, 0)

# Rodar os testes


unittest.main(argv=[''], exit=False)
````

````python
# 🔁 Teste de Integração
# Testa a sequência de chamadas: depositar -> sacar -> consultar saldo
def test_fluxo_completo():
    conta = Conta("Maria", 200)
    depositar(conta, 100)
    sacar(conta, 50)
    assert conta.consultar_saldo() == 250

# Rodar o teste
test_fluxo_completo()
````

````python
# ✅ Teste de Sistema (Uso Real)
def main():
    conta = Conta("Carlos", 500)
    print("Saldo Inicial:", conta.consultar_saldo())
    depositar(conta, 300)
    sacar(conta, 200)
    print("Saldo Final:", conta.consultar_saldo())

main()
````

````python
# 🧑‍💻 Execução com entrada de dados do usuário
def main_interativo():
    nome = input("Informe o nome do titular da conta: ")
    saldo_inicial = float(input("Informe o saldo inicial: "))
    conta = Conta(nome, saldo_inicial)

    while True:
        print("\n1 - Consultar saldo")
        print("2 - Depositar")
        print("3 - Sacar")
        print("4 - Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            print("Saldo atual:", conta.consultar_saldo())
        elif opcao == '2':
            valor = float(input("Valor a depositar: "))
            depositar(conta, valor)
        elif opcao == '3':
            valor = float(input("Valor a sacar: "))
            try:
                sacar(conta, valor)
            except ValueError as e:
                print("Erro:", e)
        elif opcao == '4':
            print("Encerrando operação.")
            break
        else:
            print("Opção inválida.")

# Descomente para executar:
main_interativo()
````

````python
# Teste da transferência
def test_transferencia():
    conta1 = Conta("Ana", 1000)
    conta2 = Conta("Lucas", 500)
    transferir(conta1, conta2, 300)
    assert conta1.consultar_saldo() == 700
    assert conta2.consultar_saldo() == 800

# Rodar o teste
test_transferencia()
````

### Objetivo:
Implementar uma funcionalidade que registre o histórico de transações da conta (depósitos e saques), e criar testes unitários e de integração para validar o comportamento.

### Requisitos:
1. Adicione um atributo `historico` na classe `Conta`.
2. Sempre que um saque ou depósito for realizado, adicione um registro no histórico.
3. Crie uma função `mostrar_historico()` que exibe a lista de transações.
4. Escreva um teste que verifique se o histórico está correto após duas operações.

### Exemplo esperado:
```python
conta.mostrar_historico()
# Saída:
# ['Depósito de R$100.0', 'Saque de R$50.0']
```


````python
class Conta:
    def __init__(self, nome, saldo):
        self.nome = nome
        self.saldo = saldo
        self.historico = []

    def consultar_saldo(self):
        return self.saldo

    def mostrar_historico(self):
        return self.historico


def sacar(conta, valor):
    conta.saldo -= valor
    conta.historico.append(f"Saque de R${valor}")


def depositar(conta, valor):
    conta.saldo += valor
    conta.historico.append(f"Depósito de R${valor}")


def transferir(conta_origem, conta_destino, valor):
    sacar(conta_origem, valor)
    depositar(conta_destino, valor)
````

````python
def test_historico():
    conta = Conta("Ana", 100)

    depositar(conta, 100)
    sacar(conta, 50)

    assert conta.mostrar_historico() == [
        "Depósito de R$100",
        "Saque de R$50"
    ]


test_historico()
````


## 💡 Desafio 3 - Limite de Saques Diários

### Objetivo:
Limitar a quantidade de saques por dia e testar esse comportamento com unit test e sistema.

### Requisitos:

2. Permita no máximo 3 saques por dia.
3. Após o terceiro saque, a função `sacar()` deve levantar um erro personalizado.

5. Teste com pelo menos 4 chamadas consecutivas da função `sacar()`.

### Dica de implementação:

### ✅ Conclusão
- Testes ajudam a prevenir erros e garantir qualidade.
- Testes unitários garantem o funcionamento isolado.
- Testes de integração validam a comunicação entre funções.
- Testes de sistema simulam o uso real da aplicação.
- Você pode tornar a aplicação interativa com `input()` no Google Colab.

````python
class LimiteDeSaquesExcedido(Exception):
    pass


class Conta:
    def __init__(self, nome, saldo):
        self.nome = nome
        self.saldo = saldo
        self.historico = []
        self.saques_diarios = 0

    def consultar_saldo(self):
        return self.saldo

    def mostrar_historico(self):
        return self.historico

    def resetar_saques(self):
        self.saques_diarios = 0


def sacar(conta, valor):
    if valor <= 0:
        raise ValueError("Valor inválido")

    if valor > conta.saldo:
        raise ValueError("Saldo insuficiente")

    if conta.saques_diarios >= 3:
        raise LimiteDeSaquesExcedido("Limite diário de saques atingido")

    conta.saldo -= valor
    conta.saques_diarios += 1
    conta.historico.append(f"Saque de R${valor}")
    return conta.saldo


def depositar(conta, valor):
    if valor <= 0:
        raise ValueError("Valor inválido")

    conta.saldo += valor
    conta.historico.append(f"Depósito de R${valor}")
    return conta.saldo


def transferir(conta_origem, conta_destino, valor):
    sacar(conta_origem, valor)
    depositar(conta_destino, valor)
````

````python
def test_saque_valido_equivalencia():
    conta = Conta("João", 100)
    resultado = sacar(conta, 50)
    print("Novo saldo esperado: 50")
    print("Novo saldo obtido:", resultado)


def test_saque_invalido_equivalencia():
    conta = Conta("João", 100)
    try:
        sacar(conta, 150)
        print("ERRO: Saque acima do saldo foi permitido!")
    except ValueError as e:
        print("Teste passou:", e)


def test_limite_saques():
    conta = Conta("João", 500)

    try:
        sacar(conta, 50)
        sacar(conta, 50)
        sacar(conta, 50)
        sacar(conta, 50)
        print("ERRO: Limite não funcionou")
    except LimiteDeSaquesExcedido as e:
        print("Teste passou:", e)


test_saque_valido_equivalencia()
test_saque_invalido_equivalencia()
test_limite_saques()
````

✅ O que é Teste de Classes de Equivalência?


O Teste de Classes de Equivalência é uma técnica de projeto de casos de teste onde os dados de entrada são divididos em classes ou grupos equivalentes, que devem se comportar da mesma forma.


A ideia é testar um valor representativo de cada grupo, reduzindo a quantidade de testes sem perder eficácia.

🔍 Por que usar?
Reduz o número de casos de teste

Aumenta a cobertura com menos esforço

Foca em representatividade, não em quantidade
